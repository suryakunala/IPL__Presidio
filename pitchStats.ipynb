```python
#dataTables
ball_by_ball=pd.read_csv('Ball_by_Ball.csv')
matchData=pd.read_csv('Match.csv')
PlayerData=pd.read_csv('Player.csv')
SeasonData=pd.read_csv('Season.csv')
TeamData=pd.read_csv('Team.csv')
```

```python
matchData.groupby()
```

```python
c=matchData.groupby('Venue_Name')
```

```python
stadiumData=[]
for stadium,index in c.groups.items():
    j,k=0,0
    for ind in index:
        a=list(matchData.loc[ind,['Toss_Decision','Win_Type']])
        #print(a)
        if a[0]=='field':
            j+=1
        if a[1]=='by wickets':
            k+=1
    
    stadiumData.append([stadium,j,k,len(index)])
    
```

```python
c.groups.items()
```

```python
stadiumData=pd.DataFrame(stadiumData,columns=['stadiumName','tossWonChoseField','chaseWon','Total'])
```

```python
stadiumData
```

```python
stadiumData['tossWonChoseField']=stadiumData['tossWonChoseField']/stadiumData['Total']
```

```python
stadiumData['chaseWon']=stadiumData['chaseWon']/stadiumData['Total']
```

```python
stadiumData.tail()
```

```python
stadiumData.to_csv('stadiumData.csv')
```
