** Read Data**

```python
data=pd.read_csv('FinalData.csv')
```

```python
data.tail()
```

```python
labels=data['Match_Winner_Id']
```

```python
train=data.drop(['Match_Winner_Id','Unnamed: 0'],axis=1)
```

```python
train.tail()
```

**Normalize Data**

```python
train['Economy']=train['Economy']/train['Economy'].max()
train['wickets']=train['wickets']/train['wickets'].max()
train['EconomyInDeathOvers']=train['EconomyInDeathOvers']/train['EconomyInDeathOvers'].max()
train['WicketsInDeathOvers']=train['WicketsInDeathOvers']/train['WicketsInDeathOvers'].max()
train['targetScore']=train['targetScore']/train['targetScore'].max()
train['requiredScore']=train['requiredScore']/train['requiredScore'].max()
```

**Split data**

```python
xTrain,xTest,yTrain,yTest=train_test_split(train,labels,test_size=0.25)
```

```python
model=LogisticRegression(multi_class="multinomial",solver="lbfgs", C=10)
model
```

```python
model.fit(train,labels)
```

```python
yPred=list(model.predict_proba(xTest))
```

```python
for val in yPred:
    print('The Probability of winning CSK when Dhoni is in Striking/nonStriking playing last over is ' + str(val[1]))
```
