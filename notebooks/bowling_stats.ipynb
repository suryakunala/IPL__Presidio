```python
ball_by_ball=pd.read_csv('Ball_by_Ball.csv')
matchData=pd.read_csv('Match.csv')
PlayerData=pd.read_csv('Player.csv')
SeasonData=pd.read_csv('Season.csv')
TeamData=pd.read_csv('Team.csv')
```

```python
ball_by_ball.tail()
```

```python
dicti=ball_by_ball.groupby('Bowler_Id')
```

```python
ball_by_ball.stats()
```

```python
bowlerData={}
for id,indices in dicti.groups.items():
    arr=[]
    for i in indices:
        arr.append(list(ball_by_ball.loc[i,['Over_Id','Ball_Id','Batsman_Scored','Extra_Runs','Player_dissimal_Id']]))
    bowlerData[id]=arr
```

```python
bowlerData
```

```python
bowlerStats=[]
for key,values in bowlerData.items():
    ballscount=len(values)
    score=0
    wickets=0
    deathoverballs=0
    deathoverwickets=0
    deathoverruns=0
    for value in values:
        try:    
            if value[0]>15:
                deathoverballs+=1
                if value[4]!=' ':
                    deathoverwickets+=1
                deathoverruns+=int(value[2][-1])
            score+=int(value[2][-1])
            if value[4]!=' ':
                wickets+=1
        except:
            pass
    econ=round(score*6/ballscount,3)
    if deathoverballs:
        deathecon=round(deathoverruns*6/deathoverballs,3)
    else:
        deathecon=econ
    bowlerStats.append([key,econ,wickets,deathecon,deathoverwickets])

#econ,wickets,deathecon,deathoverwickets
```

```python
bowlerStat=pd.DataFrame(bowlerStats,columns=['Bowler_id','econ','wickets','deathecon','deathoverwickets'])
```

```python
bowlerStat.head()
```

```python
bowling_skill=[]
for value in bowlerStats:
    bowling_skill.append(list(PlayerData.loc[value[0],['Bowling_Skill']])[-1])
```

```python
bowlerStat['Bowling_type']=bowling_skill
```

```python
bowlerStat
```

```python
bowlerStat.to_csv('bowlerStat.csv')
```
