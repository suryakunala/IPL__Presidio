```python
ball_by_ball=pd.read_csv('Ball_by_Ball.csv')
matchData=pd.read_csv('Match.csv')
PlayerData=pd.read_csv('Player.csv')
SeasonData=pd.read_csv('Season.csv')
TeamData=pd.read_csv('Team.csv')
```

```python
ballsFacedbyDhoni=list(dicti.groups[20])
```

```python
data=[]
for i in ballsFacedbyDhoni:
    data.append(list(ball_by_ball.loc[i]))
```

```python
DhoniData=pd.DataFrame(data,columns=ball_by_ball.columns)
```

```python
DhoniData.tail()
```

```python
Dhoni20Data=DhoniData[(DhoniData['Over_Id']==20) & (DhoniData['Innings_Id']==2)] 
```

```python
Dhoni20Data
```

```python
Dhoni20Data
```

```python
c=Dhoni20Data.groupby('Match_Id')
```

```python
MatchesDhoniPlayedLastOver=list(c.groups.keys())
```

```python
matchDataDhoni=[]
for val in MatchesDhoniPlayedLastOver:
    matchDataDhoni.append(list(matchData.loc[matchData[matchData['Match_Id']==val].index[-1]]))
```

```python
targetscoreData=[]
currentscoreData=[]
gro=ball_by_ball.groupby('Match_Id')
for value in MatchesDhoniPlayedLastOver:
    lis=list(gro.groups[value])
    targetScore=0
    current_score=0
    for val in lis:
        if ball_by_ball.loc[val,'Innings_Id']==1:
                s=ball_by_ball.loc[val,'Batsman_Scored']
                if s.isdigit():
                    targetScore+=int(s)
        else:
            if ball_by_ball.loc[val,'Over_Id']<20:
                s=ball_by_ball.loc[val,'Batsman_Scored']
                if s.isdigit():
                    current_score+=int(s)
    targetscoreData.append(targetScore)
    currentscoreData.append(current_score)
```

```python
requiredScore=[0]*16
for i in range(16):
    requiredScore[i]=targetscoreData[i]-currentscoreData[i]
    if requiredScore[i]<0:
        requiredScore[i]=5
```

```python
matchdataDhoni=pd.DataFrame(matchDataDhoni,columns=matchData.columns)
```

```python
matchdataDhoni.head()
```

```python
Data1=matchdataDhoni.loc[:,['Match_Id','Opponent_Team_Id','Venue_Name','Toss_Winner_Id','Toss_Decision','Win_Type','Won_By','Match_Winner_Id']]
```

```python
temp=list(Data1['Match_Id'])
bowlid=[]
for tem in temp:
    bowlid.append(Dhoni20Data.loc[Dhoni20Data[Dhoni20Data['Match_Id']==tem].index[-1],'Bowler_Id'])
```

```python
Data1['bowler_id']=bowlid
Data1.tail()
```

```python
data2=pd.read_csv('bowlerStat.csv')
```

```python
data2
```

```python
data=[]
for i in bowlid:
    data.append(list(data2.loc[data2[data2['Bowler_id']==i].index[-1],['econ','wickets','deathecon','deathoverwickets']]))
```

```python
econ=[]
wickets=[]
deathecon=[]
deathoverwickets=[]
for val in data:
    econ.append(val[0])
    wickets.append(val[1])
    deathecon.append(val[2])
    deathoverwickets.append(val[3])
Data1['Economy']=econ
Data1['wickets']=wickets
Data1['EconomyInDeathOvers']=deathecon
Data1['WicketsInDeathOvers']=deathoverwickets
```

```python
venuedata=[]
for i in range(16):
    venuedata.append(Data1.loc[i,'Venue_Name'])
```

```python
data3=pd.read_csv('stadiumData.csv')
```

```python
data=[]
for i in venuedata:
    data.append(list(data3.loc[data3[data3['stadiumName']==i].index[-1],['tossWonChoseField','chaseWon']]))
```

```python
tossWonChoseField=[]
chaseWon=[]
for val in data:
    tossWonChoseField.append(val[0])
    chaseWon.append(val[1])
Data1['tossWonChoseField']=tossWonChoseField
Data1['chaseWon']=chaseWon
```

```python
replacement_mapping_dict1={
    'field':1,
    'bat':0
}
replacement_mapping_dict2={
    'by runs':0,
    'by wickets':1
}
replacement_mapping_dict3={
    3.0:1,
    1:0,
    2:0,
    4:0,
    5:0,
    6:0,
    7:0,
    8:0,
    9:0,
    10:0,
    11:0,
    12:1,
    13:0
}
Data1['Toss_Decision'].replace(replacement_mapping_dict1,inplace=True)
Data1['Win_Type'].replace(replacement_mapping_dict2,inplace=True)
Data1['Match_Winner_Id'].replace(replacement_mapping_dict3,inplace=True)
```
